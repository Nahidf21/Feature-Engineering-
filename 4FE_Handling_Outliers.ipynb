### 5 Number Summary And Box Plot 

```python
## Minimum Maximum Median Q1 Q2 IQR 
```

```python
import numpy as np
```

```python
first_marks = [45,32,56,75,89,54,32,89,90,87,67,74,54,45,98,99,67]
minimum,Q1,median,Q3,maximum = np.quantile(first_marks,[0,0.25,0.50,0.75,1.0])
```

```python
print(minimum)
print(Q1)
print(median)
print(Q3)
print(maximum)

```

```python
IQR = Q3-Q1
IQR
```

```python
lower_fence = Q1-1.5 * (IQR)
higer_fence = Q3 + 1.5 * (IQR)
```

```python
lower_fence
```

```python
higer_fence
```

```python
import seaborn as sns 
```

```python
sns.boxenplot(first_marks)
```
